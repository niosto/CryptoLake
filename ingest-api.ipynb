```python
import requests
import json
from datetime import datetime, timezone
from pyspark.sql import Row

# 1. Fetch the data
response = requests.get("https://api.coinlore.net/api/tickers/?limit=50").json()

# 2. Extract the list of tickers
# Note: The CoinLore API returns a dict: {"data": [...], "info": {...}}
ticker_list = response['data']

# 3. Create a Spark DataFrame
df = spark.createDataFrame([Row(**ticker) for ticker in ticker_list])

# 4. Generate your timestamp for the path
utc_now = datetime.now(timezone.utc).strftime("%Y-%m-%d_%H-%M-%S")
path = f"dbfs:/bfs/tickers_{utc_now}.parquet"

# 5. Save as Parquet
df.write.mode("overwrite").parquet(path)

print(f"Data successfully saved as Parquet at: {path}")
```
